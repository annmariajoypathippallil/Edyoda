```python
#  1.Delete the elements in an linked list whose sum is equal to zero

def delete_zero_sum(head):
    if not head or not head.next:
        return head

    dummy = Node(0)
    dummy.next = head
    prefix_sum = 0
    prefix_sums = {prefix_sum: dummy}

    current = head
    while current:
        prefix_sum += current.data

        if prefix_sum in prefix_sums:
            prev = prefix_sums[prefix_sum]
            prev.next = current.next
            while prev.next != current.next:
                prev = prev.next
                del prev

        else:
            prefix_sums[prefix_sum] = current

        current = current.next

    return dummy.next



head = Node(3)
head.next = Node(4)
head.next.next = Node(-7)
head.next.next.next = Node(2)
head.next.next.next.next = Node(-3)
head.next.next.next.next.next = Node(1)
head.next.next.next.next.next.next = Node(6)
head.next.next.next.next.next.next.next = Node(-6)

print("Original Linked List:")
print_list(head)

head = delete_zero_sum(head)

print("Modified Linked List:")
print_list(head)
```

```python
# 3.Merge a linked list into another linked list at alternate positions.


class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None


def merge_lists(head1, head2):
    if not head1:
        return head2
    if not head2:
        return head1

    current1 = head1
    current2 = head2
    while current1 and current2:
        next1 = current1.next
        next2 = current2.next

        current1.next = current2
        current2.next = next1

        current1 = next1
        current2 = next2

    return head1



def create_linked_list():
    head = None
    tail = None

    while True:
        data = input("Enter a value (or 'q' to quit): ")
        if data == 'q':
            break

        new_node = Node(int(data))

        if not head:
            head = new_node
            tail = new_node
        else:
            tail.next = new_node
            tail = new_node

    return head



def print_list(head):
    current = head
    while current:
        print(current.data, end=" -> ")
        current = current.next
    print("None")


print("Create the first linked list:")
head1 = create_linked_list()

print("\nCreate the second linked list:")
head2 = create_linked_list()

print("\nFirst Linked List:")
print_list(head1)

print("\nSecond Linked List:")
print_list(head2)

head1 = merge_lists(head1, head2)

print("\nMerged Linked List:")
print_list(head1)
```

```python
# 2.Reverse a linked list in groups of given size




class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None


def reverse_linked_list_in_groups(head, k):
    current = head
    next_node = None
    prev = None
    count = 0

    while current is not None and count < k:
        next_node = current.next
        current.next = prev
        prev = current
        current = next_node
        count += 1

    if next_node is not None:
        head.next = reverse_linked_list_in_groups(next_node, k)

    return prev


def print_linked_list(head):
    current = head
    while current:
        print(current.data, end=" ")
        current = current.next
    print()


arr = input("Enter the elements of the linked list separated by spaces: ").split()
arr = [int(num) for num in arr]

head = Node(arr[0])
temp = head
for i in range(1, len(arr)):
    new_node = Node(arr[i])
    temp.next = new_node
    temp = new_node

print("Original Linked List:")
print_linked_list(head)

k = int(input("Enter the group size: "))

head = reverse_linked_list_in_groups(head, k)

print("Reversed Linked List:")
print_linked_list(head)
```

```python
#  4.In an array, Count Pairs with given sum


def count_pairs_with_sum(arr, target_sum):
    frequency = {}
    count = 0

    for num in arr:
        if num in frequency:
            frequency[num] += 1
        else:
            frequency[num] = 1

    for num in arr:
        difference = target_sum - num

        if difference in frequency:
            count += frequency[difference]

        frequency[num] -= 1

    return count // 2


arr = input("Enter the array elements separated by spaces: ").split()
arr = [int(num) for num in arr]

target_sum = int(input("Enter the target sum: "))

pairs_count = count_pairs_with_sum(arr, target_sum)
print("Number of pairs with sum", target_sum, ":", pairs_count)

```

```python
# 5.Find duplicates in an array

def find_duplicates(arr):
    seen = set()
    duplicates = {}

    for num in arr:
        if num in seen:
            if num in duplicates:
                duplicates[num] += 1
            else:
                duplicates[num] = 2
        else:
            seen.add(num)

    return duplicates

arr = input("Enter the array elements separated by spaces: ").split()
arr = [int(num) for num in arr]

duplicates = find_duplicates(arr)
duplicate_count = sum(duplicates.values())

print("Duplicate elements:")
for num, count in duplicates.items():
    print(f"{num} (Count: {count})")

print("Total duplicates:", duplicate_count)
```

```python
# 8.Reverse a string using a stack data structure

class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()

    
def reverse_string(input_string):
    stack = Stack()
    for char in input_string:
        stack.push(char)

    reversed_string = ''

    while not stack.is_empty():
        reversed_string += stack.pop()

    return reversed_string


input_string = input("Enter a string: ")

reversed_string = reverse_string(input_string)
print("Reversed string:", reversed_string)



```

```python
# 9.Evaluate a postfix expression using stack

class Stack:
    def __init__(self):
        self.items = []

    def is_empty(self):
        return len(self.items) == 0

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()


def evaluate_postfix(expression):
    stack = Stack()

    operators = {'+': lambda a, b: a + b,
                 '-': lambda a, b: a - b,
                 '*': lambda a, b: a * b,
                 '/': lambda a, b: a / b}

    for element in expression:
        if element.isdigit():
            stack.push(int(element))
        elif element in operators:
            operand2 = stack.pop()
            operand1 = stack.pop()
            operation = operators[element]
            result = operation(operand1, operand2)
            stack.push(result)

    return stack.pop()


expression = input("Enter a postfix expression: ").split()

result = evaluate_postfix(expression)
print("Result:", result)
```

```python
# 10.Implement a queue using the stack data structure

class Queue:
    def __init__(self):
        self.stack_enqueue = []
        self.stack_dequeue = []

    def is_empty(self):
        return len(self.stack_enqueue) == 0 and len(self.stack_dequeue) == 0

    def enqueue(self, item):
        self.stack_enqueue.append(item)

    def dequeue(self):
        if self.is_empty():
            return None

        if len(self.stack_dequeue) == 0:
            while len(self.stack_enqueue) > 0:
                self.stack_dequeue.append(self.stack_enqueue.pop())

        return self.stack_dequeue.pop()


queue = Queue()

values = input("Enter values to enqueue (separated by spaces): ").split()

for value in values:
    queue.enqueue(value)

print("Dequeued values:")
while not queue.is_empty():
    print(queue.dequeue())
```
