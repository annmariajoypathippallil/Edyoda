```python
import pandas as pd
```

1.importing data set

```python
# url = 'https://raw.githubusercontent.com/SR1608/Datasets/main/covid-data.csv'
data = pd.read_csv('https://raw.githubusercontent.com/SR1608/Datasets/main/covid-data.csv')
data.head(5)
```

2.High level understanding of data

```python
# a.Find no. of rows and cloumns in a dataset


num_rows, num_columns = data.shape

print(f"Number of rows: {num_rows}")
print(f"Number of columns: {num_columns}")
```

```python
# b.Data type of columns
column_data_types = data.dtypes
print("Data type of columns", column_data_types)
```

```python
# c.info of data frames
data.info()
```

```python
#c. Description of data frames
data.describe()
```

3.Low level data understanding

```python
# a.Find cound of unique values in location column

unique_location_count = data["location"].nunique()
print(f"Count of unique values in 'location' column: {unique_location_count}")
```

```python

# b.find which continent has the maximum number of frequency using values count
continent_counts = data['continent'].value_counts()
max_frequency_continent = continent_counts.idxmax()
max_frequency_count = continent_counts.max()
print(f"The continent with the maximum frequency is '{max_frequency_continent}' with a count of {max_frequency_count}.")
```

```python
# c.Find maximum and mean_value in total_cases

maximum_total_cases=data["total_cases"].max()
mean_total_cases = data["total_cases"].mean()
print(f"Maximum of 'total_cases':{maximum_total_cases}\nMean of 'total_cases':{mean_total_cases}")
```

```python
# d. find 25%,50% and 70% quartile value in 'total_deaths'

quartiles = [0.25, 0.50, 0.75]
quartile_values = data['total_deaths'].quantile(quartiles)
for q, value in zip(quartiles, quartile_values):
    print(f"{int(q * 100)}% Quartile: {value}")
```

```python
# e. find which continent has maximum human_development_index

max_hdi_idx = data.groupby('continent')['human_development_index'].idxmax()
max_hdi_continents = data.loc[max_hdi_idx]
print("Continent(s) with maximum 'human_development_index':")
for index, row in max_hdi_continents.iterrows():
    print(f"{row['continent']}: {row['human_development_index']}")
```

```python
# f.find which continent has minimum "gdp_per_capita "
continent_min_gdp = data.groupby('continent')['gdp_per_capita'].min()

# Find the continent with the minimum gdp_per_data.
continent_with_min_gdp = continent_min_gdp.idxmin()

print("Continent with Minimum GDP per Data:", continent_with_min_gdp)
```

```python
# 4. Filter the dataframe with only this columns ['continent', 'location', 'date','total_cases','total_deaths','gdp_per_ca pita',' human_development_index'] and update the data frame.

selected_columns = ['continent', 'location', 'date', 'total_cases', 'total_deaths', 'gdp_per_capita', 'human_development_index']
filtered_data = data[selected_columns]
filtered_data.to_csv('filtered_dataset.csv', index=False)
```

```python
filtered_data
```

5. Data Cleaning

```python

# a. Remove all duplicate observations
datas = data.drop_duplicates()
datas
```

```python
# b. Find missing values in all columns
missing_values = data.isnull().sum()
missing_values
```

```python
# c. Remove all observations where the continent column value is missing
datas = data.dropna(subset=['continent'])

datas
```

```python
# d. Fill all missing values with 0
datas = data.fillna(0)
datas
```

```python
# Save the cleaned DataFrame to a new CSV file
data.to_csv('cleaned_dataset.csv', index=False)
```

```python
data
```

6. Date time format:

```python
# a. Convert the "date" column to datetime format
data['date'] = pd.to_datetime(data['date'])

# b. Create a new "month" column by extracting the month data from the "date" column
data['month'] = data['date'].dt.month

# Save the updated DataFrame to a new CSV file
data.to_csv('updated_dataset.csv', index=False)

data
```

7. Data Aggregation:

```python

# a. Find the max value in all columns using groupby on "continent" column
df_groupby = data.groupby('continent').max().reset_index()

# Display the new DataFrame with max values by continent
print(df_groupby)
```

8. Feature Engineering:

```python
# Create the new feature 'total_deaths_to_total_cases'
data['total_deaths_to_total_cases'] = data['total_deaths'] / data['total_cases']

# Save the updated DataFrame to a new CSV file
data.to_csv('updated_dataset.csv', index=False)
data
```

9. Data Visualization:

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# a. Univariate analysis on 'gdp_per_capita' using a histogram
plt.figure(figsize=(10, 5))
sns.distplot(data['gdp_per_capita'], kde=False, bins=20)
plt.title('Histogram of GDP per Capita')
plt.xlabel('GDP per Capita')
plt.ylabel('Frequency')
plt.show()
```

```python
# b. Scatter plot of 'total_cases' vs 'gdp_per_capita'
plt.figure(figsize=(10, 5))
sns.scatterplot(x='gdp_per_capita', y='total_cases', data=data)
plt.title('Scatter Plot of Total Cases vs GDP per Capita')
plt.xlabel('GDP per Capita')
plt.ylabel('Total Cases')
plt.show()
```

```python
# c. Pairplot on df_groupby dataset
df_groupby = data.groupby('continent').max().reset_index()
sns.pairplot(df_groupby)
plt.show()
```

```python
# d. Bar plot of 'continent' column with 'total_cases'
plt.figure(figsize=(12, 6))
sns.catplot(x='continent', y='total_cases', kind='bar', data=data)
plt.title('Total Cases by Continent')
plt.xlabel('Continent')
plt.ylabel('Total Cases')
plt.xticks(rotation=45)
plt.show()
```
