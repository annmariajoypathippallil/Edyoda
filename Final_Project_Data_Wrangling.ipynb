3.Low level data understanding

```python
# a.Find cound of unique values in location column

unique_location_count = data["location"].nunique()
print(f"Count of unique values in 'location' column: {unique_location_count}")
```

```python

# b.find which continent has the maximum number of frequency using values count
continent_counts = data['continent'].value_counts()
max_frequency_continent = continent_counts.idxmax()
max_frequency_count = continent_counts.max()
print(f"The continent with the maximum frequency is '{max_frequency_continent}' with a count of {max_frequency_count}.")
```

```python
# c.Find maximum and mean_value in total_cases

maximum_total_cases=data["total_cases"].max()
mean_total_cases = data["total_cases"].mean()
print(f"Maximum of 'total_cases':{maximum_total_cases}\nMean of 'total_cases':{mean_total_cases}")
```

```python
# d. find 25%,50% and 70% quartile value in 'total_deaths'

quartiles = [0.25, 0.50, 0.75]
quartile_values = data['total_deaths'].quantile(quartiles)
for q, value in zip(quartiles, quartile_values):
    print(f"{int(q * 100)}% Quartile: {value}")
```

```python
# e. find which continent has maximum human_development_index

max_hdi_idx = data.groupby('continent')['human_development_index'].idxmax()
max_hdi_continents = data.loc[max_hdi_idx]
print("Continent(s) with maximum 'human_development_index':")
for index, row in max_hdi_continents.iterrows():
    print(f"{row['continent']}: {row['human_development_index']}")
```
